```python
############## for success probility 3/5
# Import the necessary modules
import random
import pandas as pd
# Setting the success probability
p = 3/5
# Setting the number of trials
trials = 500
# Setting the initial number of successes
successes = 0

# Poisson trials
for _ in range(trials):
  r = random.uniform(0, 1)
  # Check if the random number is less than the success probability
  if (r < p):
    #increment the number of successes
    successes += 1

# Print the result
print(f"Number of successes: {successes}")
```

```python
###############number of trials =500 for success probability 3/5
# Importing the necessary modules
import random
import matplotlib.pyplot as plt
# Setting the success probability
p = 3/5
# Setting the number of trials
trials = 500
# Setting the initial number of successes
successes = 0

# store the results in list
equity_curve = []
learning_curve = []

#  Poisson trials
for i in range(trials):
  # Generate a random number between 0 and 1
  randomNumber = random.uniform(0, 1)
  # Check if the random number is less than the success probability
  if randomNumber < p:
    # increment the number of successes
    successes += 1
  # Compute the equity curve value for this trial
  equity_curve.append(successes / (i+1))
  # Compute the learning curve value for this trial
  learning_curve.append(successes / trials)

# Plotting the equity curve
plt.plot(equity_curve, c="red")
plt.xlabel("Number of Trial")
plt.ylabel("Equity")
plt.show()

# Plotting the learning curve
plt.plot(learning_curve)
plt.xlabel("Number of Trial")
plt.ylabel("Learning")
plt.show()

```

```python
################## success probability = t mod 2, no. of trials =500
# Import the necessary modules
import random
import matplotlib.pyplot as plt

# Set the number of trials
trials = 500

# Set the initial number of successes
successes = 0

# Initialize lists to store the results
equity_curve = []
learning_curve = []

# Run the Poisson trials
for i in range(trials):
  # Compute the success probability for this trial
  p = i % 2
  
  # Generate a random number between 0 and 1
  r = random.uniform(0, 1)
  # Check if the random number is less than the success probability
  if r < p:
    # If so, increment the number of successes
    successes += 1
  
  # Compute the equity curve value for this trial
  equity_curve.append(successes / (i+1))

  # Compute the learning curve value for this trial
  learning_curve.append(successes / trials)

# Print the result
print(f"Number of successes: {successes}")
# Plot the equity curve
plt.plot(equity_curve, c="green")
plt.xlabel("Total Number of trials")
plt.ylabel("Equity")
plt.show()

# Plot the learning curve
plt.plot(learning_curve)
plt.xlabel("Total Number of Trials")
plt.ylabel("Learning")
plt.show()
```

```python
# Import the necessary modules
import random
import matplotlib.pyplot as plt

# Set the success probability
p = 3/5

# Set the number of trials
trials = 100

# Set the initial number of successes
successes = 0

# Initialize lists to store the results
equity_curve = []
learning_curve = []

# Run the Poisson trials
for i in range(trials):
  # Generate a random number between 0 and 1
  r = random.uniform(0, 1)
  
  # Add a small random value to the success probability to break ties
  tie_breaker = random.uniform(-0.01, 0.01)
  
  # Check if the random number is less than or equal to the success probability
  # with the tie-breaker added
  if r <= p + tie_breaker:
    # If so, increment the number of successes
    successes += 1
  
  # Compute the equity curve value for this trial
  equity_curve.append(successes / (i+1))

  # Compute the learning curve value for this trial
  learning_curve.append(successes / trials)

# Plot the equity curve
plt.plot(equity_curve)
plt.xlabel("Trial number")
plt.ylabel("Equity")
plt.show()

# Plot the learning curve
plt.plot(learning_curve)
plt.xlabel("Trial number")
plt.ylabel("Learning")
plt.show()
```

```python
###################### when success probability = [log t/log 2 ] * mod 2 , no. of trials =500
# Import the necessary modules
import random
import matplotlib.pyplot as plt
import math

# Set the number of trials
trials = 500

# Set the initial number of successes
successes = 0

# Initialize lists to store the results
equity_curve = []
learning_curve = []

# Run the Poisson trials
for i in range(trials):
  # Compute the success probability for this trial
  p = (math.log(i+1) / math.log(2)) % 2
  
  # Generate a random number between 0 and 1
  r = random.uniform(0, 1)
  # Check if the random number is less than the success probability
  if r < p:
    # If so, increment the number of successes
    successes += 1
  
  # Compute the equity curve value for this trial
  equity_curve.append(successes / (i+1))

  # Compute the learning curve value for this trial
  learning_curve.append(successes / trials)

# Print the result
print(f"Number of successes: {successes}")
# Plot the equity curve
plt.plot(equity_curve,c="purple")
plt.xlabel("Trial number")
plt.ylabel("Equity")
plt.show()

# Plot the learning curve
plt.plot(learning_curve)
plt.xlabel("Trial number")
plt.ylabel("Learning")
plt.show()
```

```python
############################ case where samples from the H action and the T action were equal for a particular round : introducing a tie factor
# Import the necessary modules
import random
import matplotlib.pyplot as plt

# Set the success probability
p = 3/5

# Set the number of trials
trials = 500

# Set the initial number of successes
successes = 0

# Initialize lists to store the results
equity_curve = []
learning_curve = []

# Run the Poisson trials
for i in range(trials):
  # Generate a random number between 0 and 1
  randomNumber = random.uniform(0, 1)
  # Adding a small random value to the success probability to break ties
  tie_break = random.uniform(-0.05, 0.05)
  
  # Check if the random number is less than or equal to the success probability with the tie-breaker added
  if randomNumber <= p + tie_break:
    # If so, increment the number of successes
    successes += 1
  
  # Compute the equity curve value for this trial
  equity_curve.append(successes / (i+1))

  # Compute the learning curve value for this trial
  learning_curve.append(successes / trials)

# Plot the equity curve
plt.plot(equity_curve)
plt.xlabel("Trial number")
plt.ylabel("Equity")
plt.show()

# Plot the learning curve
plt.plot(learning_curve)
plt.xlabel("Trial number")
plt.ylabel("Learning")
plt.show()
```
